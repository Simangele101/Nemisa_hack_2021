```python
time_series = ['200909', '200912', '201003', '201006',
       '201009', '201012', '201103', '201106', '201109', '201112', '201203',
       '201206', '201209', '201212', '201303', '201306', '201309', '201312',
       '201403', '201406', '201409', '201412', '201503', '201506', '201509',
       '201512', '201603', '201606', '201609', '201612', '201703', '201706',
       '201709', '201712', '201803', '201806', '201809', '201812', '201903',
       '201906', '201909', '201912', '202003', '202006', '202009']
date = ['Industry description']
for i in time_series:
    date.append(i[:4]+'-'+i[-2:]) 
no_employees_per_sector.columns = date
```

```python
# Pivoting the data
df = no_employees_per_sector.copy()
df = df.T
```

```python
df.head()
```

```python
# Changing the column header
new_header =  df.iloc[0]
df_2 = df[1:]
df.columns = new_header
df.drop('Industry description', inplace = True)
```

```python
df.columns
```

```python
eco_sectors = pd.DataFrame(df[['Mining and quarrying','Manufacturing',
                       'Electricity, gas and water supply','Construction','Wholesale, retail and motor  trade; hotels and restaurnts',
                      'Transport, storage and communication','Financial intermediation, insurance, real estate and business services'
                      ,'Community, social and personal servcies',]])

eco_sectors.dropna(inplace = True)
```

```python
# change the index to datetime
eco_sectors.index = pd.to_datetime(eco_sectors.index)
```

```python
for i in eco_sectors.columns:
    print(i)
    eco_sectors[str(i)].plot(figsize=(15, 6))
    plt.title('')
    plt.show()
```

```python
df2 = pd.DataFrame(eco_sectors['Mining and quarrying'])
df2['quarter'] = df2.index
df2.reset_index(drop=True, inplace = True)
# reorder the columns
df2 = df2[['quarter','Mining and quarrying']]
df2['quarter'] = df2['quarter'].astype('int64')
```

```python
# creating X and y metrics
X = df2['quarter'].values.reshape(-1,1)
y = df2['Mining and quarrying']
```

```python
# splitting the data into train set and test set
X_train, X_test, y_train, y_test = train_test_split(X,y, shuffle=False, test_size =0.3)

lm = LinearRegression()
lm.fit(X_train, y_train)
```

## Evaluation

```python
# Function that calculates the root mean squared error
def rmse(y_test, y_prediction):
    '''
    This function calculates the root mean squared error which
    we will be using to check  the accuracy of our models
    Params
    ------------
    y_test : The actual y values
    y_prediction : The predicted y values we get from out model
    '''
    result = np.sqrt(mean_squared_error(y_test, y_prediction))
    return result
```

```python
# Evaluation
rmse(y_test,lm.predict(X_test))
```

```python
def func(string):
    quarter_ls = [string]
    new_df = pd.DataFrame(quarter_ls)
    result = lm.predict([pd.to_datetime(new_df[0]).astype('int64')])
    return round(result[0],)
```
